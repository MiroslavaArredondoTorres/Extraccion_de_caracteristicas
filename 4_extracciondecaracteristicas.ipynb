```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
#Carga desde un archivo .xlsx sin indice
micro_retailer=pd.read_excel('C:/Users/Miroslava/Documents/clase analitica de datos/Analitica-de-datos/extraccion de caracteristicas/microretailer_mit_lift_lab.xlsx')
```

```python
#Verificamos información del DataFrame
micro_retailer.info()
```

```python
micro_retailer.head()
```

```python
#Limitar numero máximo de columnas
pd.options.display.max_rows = None
```

```python
#Obtengo un analis univariado de las variables categóricas
freq_tbl(Micro_Retailer)
```

```python
#Obtengo un análisis univariado de una variable categórica en específico
table = freq_tbl(micro_retailer['214_customer_relationship_tools'])
table
```

```python
#Obtengo un filtro de los valores más relevantes de las variables categóricas seleccionadas más significativas
Filter = table[table['frequency']>1]
Filter
```

```python
#Ajusto el indice de mi DataFrame
Filter_index=Filter.set_index('214_customer_relationship_tools')
Filter_index
```

```python
#Elimine las columnas de "percentage" y "cumulative_perc"

Filter_index_frequency= Filter_index.drop(['percentage', 'cumulative_perc'], axis=1)
Filter_index_frequency
```

```python
#Realizamos gráfico de barras del dataframe filtrado
Filter_index_frequency.plot(kind='bar', width=0.8, figsize=(10,4))
plt.title('Herramientas  relación con los clientes')
plt.xlabel('Tipo Herramienta')
plt.ylabel('Frecuencia')
```

```python
#Realizamos grafico de dispersión del DataFrame filtrado
Filter_index.plot("frequency", "cumulative_perc", kind="scatter", figsize=(10,5))
```

```python
#Realizamos grafico de área del dataframe filtrado
Filter_index_frequency.plot(kind='area', figsize=(10,4), alpha= 0.5)
```

```python
#Realizamos grafico de pastel del dataframe filtrado
Filter_index["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f %%")
```

```python
#Realizamos grafico hexagonal del dataframe filtrado, alternativo al scatter plot
Filter_index.plot.hexbin("frequency", "cumulative_perc", figsize=(10,5))
```
